```python
import numpy as np
import pandas as pd
import re
import pandas2arff as converter
```

```python
data = pd.read_csv('./badges2.csv', header=None)
data.shape
```

```python
data.head()
```

```python
del data[1]
del data[2]
del data[3]
del data[4]
del data[5]
del data[6]
del data[7]
del data[8]
del data[9]
del data[10]
data.head()
```

```python
def construct(row):
    allCharacters = list(re.sub("[^a-z]", "", row.lower()))
    bins = np.zeros(26, dtype=int)
    asciia = ord("a")
                         
    for c in allCharacters:
        index = ord(c) - asciia
        bins[index] += 1
        
    return pd.Series(bins)

splitted = data[0].apply(lambda row: construct(row))
splitted['label'] = data.iloc[:,1]
splitted.head()
```

```python
splitted.to_csv("badges2New.arff")
```
