```python
import sys

if 'pandas2arff' in sys.modules:  
    del sys.modules["pandas2arff"]

from pandas2arff import *
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
movies.head()
```

```python
ratings = pd.read_csv("ratings.csv")
ratings.head()
```

```python
total_votes = {}
total_ratings = {}

for index, row in ratings.iterrows():
    movieId = int(row.movieId)
    rating = float(row.rating)
    
    if movieId not in total_ratings:
        total_ratings[movieId] = 0
        total_votes[movieId] = 0
    
    total_ratings[movieId] += rating
    total_votes[movieId] += 1    
```

```python
def get_rating(row):
    movieId = row.movieId
    if movieId in total_ratings:
        row.rating = int(round(2 * (total_ratings[movieId] / total_votes[movieId])))
    return row

movies["rating"] = 0
movies = movies.apply(get_rating, axis=1)
```

```python
del movies["movieId"]
del movies["genres"]
```
