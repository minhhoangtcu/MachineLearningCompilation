## Problems faced by people with Sensorineural Hearing Loss

### Decreased audibility

People with a mild or moderate loss are more likely to hear some sounds and not others. In particular, the softer phonemes, which are usually consonants, may simply not be heard. For example, the sequence of sounds *i e a ar*, might have originated as *pick the black harp*, but could be heard as *kick the cat hard*. To make sounds audible, hearing aids have to provide amplification.

For example, a hearing loss caused all frequencies (and therefore all formants) above 700 Hz to be inaudible. The two sound below *oo* and *ee* have the similar first formants. However, their second formants are different, making the two sounds indifferentiable. Although both sounds could be detected, the similarity of their first formants would make them sound almost identical.

Note: formants are "the spectral peaks of the sound spectrum. [Wiki](https://en.wikipedia.org/wiki/Formant)

```python
x_oo, fs_oo = librosa.load('./data/oo.wav', sr=44100)
x_ee, fs_ee = librosa.load('./data/ee.wav', sr=44100)

samples_oo = x_oo.shape[0]
samples_ee = x_ee.shape[0]

ipd.Audio(x_oo, rate=fs_oo)
```

```python
ipd.Audio(x_ee, rate=fs_ee)
```

```python
X_oo = scipy.fft(x_oo)
X_ee = scipy.fft(x_ee)

X_mag_oo = numpy.absolute(X_oo)
X_mag_ee = numpy.absolute(X_ee)

f_oo = numpy.linspace(0, fs_oo, samples_oo) # split the space into n sections, for n be number of samples
f_ee = numpy.linspace(0, fs_ee, samples_ee)

tenth_oo = int(samples_oo/10) # ignore the high freq noise
tenth_ee = int(samples_ee/10)

plt.figure(1)
plt.subplot(211)
plt.plot(f_oo[:tenth_oo], X_mag_oo[:tenth_oo]) # magnitude spectrum

plt.subplot(212)
plt.plot(f_ee[:tenth_ee], X_mag_ee[:tenth_ee])

plt.xlabel('Frequency (Hz)')
plt.show()
```

To help overcome this difficulty, a hearing aid has to provide more amplification for frequencies where speech has the weakest components and where hearing loss is the greatest (i.e. usually the high frequencies).

### Decreased dynamic range

As implied above, soft sounds can be made audible merely by amplifying them. Unfortunately, it is not appropriate to amplify all sounds by the amount needed to make soft sounds audible.

If the sounds in the environment are to fit within a person’s dynamic range, a hearing aid must give more amplification to weak sounds than it does to intense sounds. This squashing of a large dynamic range of levels in the environment into a smaller range of levels at the output of the hearing aid is called **compression**. In essence, a compressor is nothing more than an amplifier that automatically turns itself down as the sound gets stronger.
