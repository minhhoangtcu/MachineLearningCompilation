### Visualize data

```python
def getDatumImg(row):
    """
    Return an image from a single np array with shape 1x400
    """
    width, height = 20, 20
    square = row.reshape(width,height)
    return square

def displayData(X):
    """
    Display a 10x10 grid of images from 100 random rows from X
    """
    width, height = 20, 20
    nrows, ncols = 10, 10
    
    indices_to_display = random.sample(range(X.shape[0]), nrows*ncols)
    big_picture = np.zeros((height*nrows, width*ncols))
    
    irow, icol = 0, 0
    for idx in indices_to_display:
        if icol == ncols:
            irow += 1
            icol  = 0
         # ignore the first column and transpose the output to compensate for assigining into big_pic
        iimg = getDatumImg(X[idx,1:]).T
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
        
    fig = plt.figure(figsize=(6,6))
    img = scipy.misc.toimage(big_picture)
    plt.imshow(img,cmap = cm.Greys_r)
```

### Vectorizing Logistic Regression

#### Vectorizing the cost function

Prediction with logistic regression:

$$h_\theta(x) = g(\theta_0 + \theta_1x_1 + \theta_2x_1^2 + \theta_3x_1^2x_2 + \theta_4x_1^2x_2^2 + \theta_5x_1^2x_2^3 + ...)$$

for $g(z)$ is the sigmoid function.

$$g(z) = \frac{1}{1 + e^{-z}}$$

```python
def sigmoid(z):
    return(1.0 / (1 + np.exp(-z)))
```

```python
# Logistic regression hypothesis. 
def h(theta, X):
    return sigmoid(np.dot(X, theta))

def computeCost(theta, X, y):
    m = y.size
    pred = h(theta, X)
    
    term1 = -np.dot(y.T, np.log(pred))
    term2 = -np.dot((1 - y).T, np.log(1 - pred))
    J = (term1 +  term2) / m
    return J
```

#### Vectorizing the gradient

```python
def lrgradientReg(theta, Lambda, X, y):
    m = y.size
    h = sigmoid(X.dot(theta.reshape(-1,1)))
      
    grad = (1/m)*X.T.dot(h-y) + (Lambda/m)*np.r_[[[0]],theta[1:].reshape(-1,1)]
        
    return(grad.flatten())
```

#### Vectorizing regularized logistic regression

```python
def lrcostFunctionReg(theta, X, y, Lambda):
    m = y.size
    
    cost = computeCost(theta, X, y)
    term3 = np.sum(np.square(theta[1:])) # ignore theta0
    J = cost + Lambda/(2*m)*term3
    
    return J
```
